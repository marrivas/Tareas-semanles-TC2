### Transferencia propuesta

$ T(s) = \frac{s - p}{s + p} $  

```python
from splane import pzmap, bodePlot, GroupDelay
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt

# Transferencia

p = 1

numerador   = [1, -p] 
denominador = [1 , p]

my_tf = TransferFunction(numerador, denominador)

bodePlot(my_tf,1)

pzmap(my_tf,2)

GroupDelay(my_tf,3)
```

## Calculo del retardo de grupo

$ T(s) = \frac{s - p}{s + p} $  
$ T(jw) = \frac{jw - p}{jw + p} $  
$ \phi(w) = \sum{\phi_z} - \sum{\phi_p} $  
$ \phi(w) = \sum{arctan(\frac{w}{-p})} - \sum{arctan(\frac{w}{p})} $  

$ \frac{d}{dw} arctan(\frac{w}{-p}) = \frac{1}{1 + \frac{w^2}{p^2}} \cdot \frac{d}{dw}\frac{w}{-p} $  
$ \frac{d}{dw} arctan(\frac{w}{-p}) = \frac{p^2}{p^2 + w^2} \cdot \frac{1}{-p} $  
$ \frac{d}{dw} arctan(\frac{w}{-p}) = \frac{-p}{p^2 + w^2} $  

$ \frac{d}{dw} arctan(\frac{w}{p}) = \frac{1}{1 + \frac{w^2}{p^2}} \cdot \frac{d}{dw}\frac{w}{p} $  
$ \frac{d}{dw} arctan(\frac{w}{p}) = \frac{p^2}{p^2 + w^2} \cdot \frac{1}{p} $  
$ \frac{d}{dw} arctan(\frac{w}{p}) = \frac{p}{p^2 + w^2} $  

$ \frac{d}{dw} \phi(w) = \frac{-p}{p^2 + w^2} - \frac{p}{p^2 + w^2} $  
$ \frac{d}{dw} \phi(w) = \frac{-2p}{p^2 + w^2} $  
$ \frac{d}{dw} \phi(w) = \frac{-2}{1 + w^2} $  

$ \tau(w) = - \frac{d}{dw} \phi(w) $  
$ \tau(w) = \frac{2}{1 + w^2} $  

#### Necesito rotación de fase de 15° en W=1


Fase: Siempre arranco en $\pi$ por el cero
$w = 0$ -> Aporte Ceros - Aporte polos $= 180° - 0° = 180°$  
$w = 1$ -> $ = 172,5° - 7,5° = 165° $  
165° es la fase necesaria para la rotación de 15° pedida

$ \alpha = 7,5° $  
$ tg(\alpha) = \frac{1}{|p|} $  
$ |p| = \frac{1}{tg(7,5°)} = 7,595754113 $  

```python
# Transferencia

p = 7.595754113

numerador   = [1, -p] 
denominador = [1 , p]

my_tf = TransferFunction(numerador, denominador)

bodePlot(my_tf,1)

pzmap(my_tf,2)

GroupDelay(my_tf,3)
```

$ wo = 1 -> \frac{1}{LC} = 1 $
