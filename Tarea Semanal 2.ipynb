### Hallar la transferencia T=VoVi en función de ωo y Q.

#### Nodos:

$V_A = 0$    $V_B = 0$   $V_C = 0$  
1) $V_A \cdot (G_1+G_2+G_3+SC) - V_i \cdot G_1 - V_o \cdot G_3 - V_1 \cdot (G_2 + SC) = 0$  
2) $V_B \cdot (G_3+SC) - V_i \cdot G_3 - V_2 \cdot SC = 0$   
3) $V_C \cdot (G_4+G_4) - V_o \cdot G_4 - V_2 \cdot G_4 = 0$  

1) $V_A \cdot (G_1+G_2+G_3+SC) - V_i \cdot G_1 - V_o \cdot G_3 - V_1 \cdot (G_2 + SC) = 0$  
2) $V_1 = -V_2 \cdot \frac{SC}{G_3}$   
3) $V_2 = - V_o$  

4) 3)->2) $V_1 = V_o \cdot \frac{SC}{G_3}$  
4) -> 1) $V_A \cdot (G_1+G_2+G_3+SC) - V_i \cdot G_1 - V_o \cdot G_3 - V_o \cdot \frac{SC}{G_3} \cdot (G_2 + SC) = 0$   
$- V_i \cdot G_1 = V_o \cdot (G_3 + \frac{SC}{G_3} \cdot (G_2 + SC)) $  
$\frac{V_o}{V_i} = \frac{- G_1}{G_3 + SC \cdot \frac{G_2}{G_3} + S^2C^2 \cdot \frac{1}{G_3}}$  
$\frac{V_o}{V_i} = \frac{\frac{- G_1 \cdot G_3}{C^2}}{S^2 + S \cdot \frac{G_2}{C} + \frac{G_3^2}{C^2}}$   
$\frac{V_o}{V_i} = \frac{\frac{1}{R_3^2 C^2} \cdot \frac{-R_3}{R_1}}{S^2 + S \cdot \frac{1}{R_2C} + \frac{1}{R_3^2C^2}}$   

$W_o^2 = \frac{1}{R_3^2 C^2}$  $\frac{W_o}{Q} = \frac{1}{R_2C}$  $K = \frac{-R_3}{R_1}$  
$T(s) = \frac{K \cdot W_o^2}{s^2 + s \cdot \frac{W_o}{Q} + W_o^2}$  

### Obtener el valor de los componentes del circuito de forma tal que ωo=1 y Q=3

$W_o = 1$ -> $R_3 = 100 k\Omega$ $C = 100\mu F$  
$Q = R_2 \cdot C = 3$ -> $R_2 = 300 k\Omega$   

### Ajustar el valor de R1 de forma tal que |T(0)|=20dB.

$T(0) = K$  -> $20 dB$ -> $K = 10$  -> $R_1 = 10 k\Omega$

```python
# -*- coding: utf-8 -*-
"""
Created on Sun Apr 10 12:32:41 2022

@author: marti
"""

from splane import pzmap, bodePlot
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt

w0 = 1
Q = 3
K = 10

my_tf = TransferFunction([-K*w0**2],[1,w0/Q,w0**2])

bodePlot(my_tf,1)

pzmap(my_tf,2)
```
